## Molecular speed
The kinetic molecular theory of gases starts with the concept of molecular speed and the average and relative speed of molecules in a sample of gas. The kinetic molecular theory is an important first step in understanding the average properties of collections of non-interacting molecules. These properties include temperature, pressure, and volume. Let's try to understand this better using some python code and we will use the plotting capability to demonstrate some of the science behind the equations.
###Maxwell-Boltzmann Distribution of Molecular Speeds
$f(v_{x})dv_{x}=\sqrt{\frac{m}{2\pi k_{b}T}}e^{\frac{-m v_{x}^{2}}{2 k_{b} T}}dv_{x}$


$\frac{dN}{N}=4\pi(\frac{M}{2 \pi R T})^{\frac{3}{2}}c^{2}e^{-\frac{M v_{x}^{2}}{2RT}}dv_{x}$

```python
import math
import numpy as np # numerical Python array library
import matplotlib.pyplot as plt  # Plotting library
%matplotlib inline
def f(vx, Mass, T):
  Mass/=1000 # Convert from grams to kilogram units of R (gas constant)
  return 4*math.pi*np.power(Mass/(2*math.pi*value.R*T),3/2)*vx**2*np.exp(-Mass*vx**2/(2*value.R*T))
vx=np.arange(0,2000,.1)
print(vx)
y=f(vx,12,300)
plt.plot(vx,y)
plt.xlabel("x velocity")
plt.ylabel("Probability")
plt.savefig('Maxwell_Boltzmann.png')
```

#Defining and plotting functions

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```
